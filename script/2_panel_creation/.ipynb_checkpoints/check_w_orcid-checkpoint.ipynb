```python
(test_w_inst
.withColumn('n_appearances', func.count('work_id').over(Window.partitionBy('author_id', 'inst_id')))
.withColumn('n_appearances_ctry', func.count('work_id').over(Window.partitionBy('author_id','country')))
.withColumn('inst_id', func.when( (func.col('n_appearances')>1) #| (func.col('n_appearances_ctry') >1)
                                 , func.col('inst_id')))
 .withColumn('parent', func.regexp_replace(func.col('parent'), '\}\,\{','\}\;\{'))
 .withColumn('parent', func.split(func.col('parent'), ';'))
 .withColumn('parent', func.transform(func.col('parent'), lambda x: func.from_json(x, func.schema_of_json('{"parent_id":"","parent_name":"","parent_type":""}'))))
 .withColumn('all_parent_ids', func.transform(func.array_distinct(func.flatten(func.collect_set(func.col('parent.parent_id')).over(Window.partitionBy('publication_year')))), lambda x: func.regexp_replace(x, 'https://openalex.org/', '')))
 .withColumn('test_drop', (func.array_contains(func.col('all_parent_ids'), func.col('inst_id'))).cast('int'))
 #.filter( (func.col('test_drop')==0) | (func.col('test_drop').isNull()))
 .select('author_id','publication_year', func.collect_set(func.col('inst_name')).over(Window.partitionBy('author_id','publication_year')).alias('insts'),
         func.max('test_drop').over(Window.partitionBy('author_id','publication_year')).alias('test_drop')).distinct().sort('author_id','publication_year','inst_name')
).show(1000, truncate = False)
```
