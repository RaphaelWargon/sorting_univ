```python
au_period = (au_inst_period
             .select('author_id','year','inst_id')
             .join(inst
                   .select('inst_id', func.when(func.col('merged_inst_id').isNull(), func.col('inst_id')).otherwise(func.col('merged_inst_id')).alias('merged_inst_id'))
                   .unionAll(inst_UK.select('inst_id',func.col('inst_id').alias('merged_inst_id'))), 
                   on = 'inst_id')
                   
             .groupBy('author_id', 'year')
             .agg(func.collect_set('merged_inst_id').alias('merged_inst_set_this_year'),
                 func.collect_set('inst_id').alias('inst_set_this_year')
                 )
             .withColumn('merged_inst_set_this_year', func.array_sort(func.col('merged_inst_set_this_year')))
             .withColumn('inst_set_this_year', func.array_sort(func.col('inst_set_this_year')))
             .withColumn('lag_inst_set_this_year', func.lag('inst_set_this_year').over(Window.partitionBy('author_id').orderBy('year')))
             .withColumn('lag_merged_inst_set_this_year', func.lag('merged_inst_set_this_year').over(Window.partitionBy('author_id').orderBy('year')))

             .withColumn('new_period', (
                                          (func.col('lag_inst_set_this_year')!=func.col('inst_set_this_year')) 
                                        | (func.col('lag_inst_set_this_year').isNull()) 
                                        ).cast('int') )
            .withColumn('period_total', func.sum('new_period').over(Window.partitionBy('author_id').orderBy('year')))
                     )
au_period.show()
```

```python
from itertools import product
au_level_treatments = (au_inst_period_smoothed_w_grant
                       .filter(func.col('country')=="FR")
                       .filter(func.col('year').between(2003,2020))
                       .withColumn('control', func.when( ( (func.col('acces_rce') == 0 ) | (func.col('acces_rce').isNull()))
                                                        & ( (func.col('date_first_idex') == 0 ) | (func.col('date_first_idex').isNull()))
                                                        & ( (func.col('fusion_date') == 0 ) | (func.col('fusion_date').isNull())),
                                                        func.col('year') ).otherwise(0))
                       .groupBy('author_id','acces_rce','date_first_idex','fusion_date', 'inst_id','inst_set_this_year')
                       .agg(func.array_sort(func.collect_set('year')).alias('all_y_in_inst'),
                            func.collect_set('control').alias('control')
                           )
                       .withColumn('control', func.explode(func.col('control')))
                       .select('author_id',
                               *[ func.expr('CASE WHEN ' + ' AND '.join([
                                   'ARRAY_CONTAINS((all_y_in_inst), (' + treat + ')-' + str(i) + ')' for i in range(y[0],y[1])]  ) + 'THEN ' + treat + ' ELSE 0 END')
                                 .alias(treat + '_' + str(y[0]) + '_' + str(y[1]) + 'y')
                                 for y,treat in product(product([0,2],[1,3,5]), ['acces_rce','date_first_idex','fusion_date','control'] ) if y[0]<=y[1]
                               ]

                              )
                       .groupBy('author_id')
                       .agg(*[ func.min(func.when(func.col(treat + '_' +  str(y[0])+'_' + str(y[1])  + 'y')!=0, 
                                                  func.col(treat + '_' +  str(y[0])+'_' + str(y[1])  + 'y')))
                                 .alias(treat + '_' + str(y[0])+'_' + str(y[1]) + 'y')
                                 for y,treat in product(product([0,2],[1,3,5]), ['acces_rce','date_first_idex','fusion_date','control']) if y[0]<=y[1]
                               ],

                            *[ func.when(func.size(func.array_remove(func.collect_set(func.col(treat + '_' +  str(y[0])+'_' + str(y[1])  + 'y').cast('int')), 0))==1, 
                                         func.min(func.col(treat + '_'+  str(y[0])+'_' + str(y[1]) + 'y'))).otherwise(0)
                                 .alias(treat + '_' +  str(y[0])+'_' + str(y[1])  + 'y_restr')
                                for y,treat in product(product([0,2],[1,3,5]), ['acces_rce','date_first_idex','fusion_date','control']) if y[0]<=y[1]
                               ]

                           )
                       .fillna(0, subset = [treat + '_' +  str(y[0])+'_' + str(y[1]) + 'y_restr'
                                for y,treat in product(product([0,2],[1,3,5]), ['acces_rce','date_first_idex','fusion_date','control']) if y[0]<=y[1]
                                           ]+
                               [treat + '_' +  str(y[0])+'_' + str(y[1]) + 'y'
                                for y,treat in product(product([0,2],[1,3,5]), ['acces_rce','date_first_idex','fusion_date','control']) if y[0]<=y[1]
                               ])
                      


                      )
au_level_treatments.cache()
au_level_treatments.show(truncate = False)
                       
```
