# Specify parameters

```python
#### CD production function:
alpha = 0.75
beta = 0.5
gamma = 0.5
#### Cost parameter for signal
k = 0.1
#### Total amount of transfers
T = 1

#### convergence to true distribution
n_convergence = 2
```

```python
def output(t,th,e):
    return( (t**alpha ) * (th ** beta) * (e ** gamma) )
```

```python
def p(n):
    return( (n-1)/(n+n_convergence) )
```

```python
plt.plot([n for n in range(1,50)], [p(n) for n in range(1, 50)])
```

```python
## Uniform theta distribution
uniform_grid = sorted(np.random.uniform(0,1,2000)) 
def f_unif(th):
    return 1/len(uniform_grid)
```

```python
## Exoponential theta distribution

lambda_ = 1/2
exp_grid = sorted(np.random.exponential(lambda_,1000)) 
def f_exp(th):
    return lambda_ *np.exp(-lambda_ *th)
```

# Equilibrium functions

```python
def no_effort_t0(t_1, T_i, f_i, n, grid):
    E_theta = sum(th * f_i(th) for th in grid)
    boundary_t0 = T_i - E_theta * t_1 * p(n)
    return(boundary_t0)
```

```python
def opt_t0(t_1, T_i, f_i, n, grid):
    d= (alpha * (1-p(n)) - k * gamma * p(n))
    E_theta = sum(th * f_i(th) for th in grid)

    det = (
        alpha**2 * t_1**2
        + (2 * k * gamma**2 / d) * t_1 * (
            (alpha + gamma) * T_i
            - (alpha + gamma) * E_theta * t_1 * p(n) 
            - t_1 * d / (2 * k)
        )
    )
    
    num = (- alpha * t_1 + np.sqrt(det) ) * d
    opt_t0 = ( num/(k * gamma ** 2))
    return(max(opt_t0, 0))
```

```python
def theta_0(t_1, T_i, f_i, n, grid):
    if t_1 >0:
        theta0 = k * gamma * opt_t0(t_1, T_i, f_i, n, grid)/ (t_1 *(alpha * (1-p(n)) - k * gamma * p(n)) )
        return( min(max(0,  theta0), 1) )
    else:
        return(1)
```

```python
def s(th,t_1, T_i, f_i, n, grid):
    if t_1 != 0 and theta_0(t_1, T_i, f_i, n, grid) <1:
        opt_s = 1/ ( (alpha + gamma) * (1- p(n)) )*( th * (alpha * (1-p(n)) - k * gamma * p(n))/k - gamma * opt_t0(t_1, T_i, f_i, n, grid) / t_1 )
        return(max([opt_s, 0]))
    else:
        return(0)
```

```python
def max_t1(T_i, f_i, n, grid):
    d= (alpha * (1-p(n)) - k * gamma * p(n))
    E_theta = sum(th * f_i(th) for th in grid)
    denom = (alpha + gamma) * E_theta * p(n) + d / (2 * k)
    max_value = (alpha + gamma) * T_i / denom
    return( max_value)
```

```python
def t_af(th,t_1, T_i, f_i, n, grid):
    if t_1 != 0 and theta_0(t_1, T_i, f_i, n, grid) <1:
        t_0 = opt_t0(t_1, T_i, f_i, n, grid) 
        opt_s =  s(th,t_1, T_i, f_i, n, grid)
        opt_t =  (t_0 + t_1 * ( p(n)*th + (1-p(n) ) * opt_s ) )
    else:
        t_0 = no_effort_t0(t_1, T_i, f_i, n, grid)
        opt_t = (t_0 + t_1 * p(n) * th ) 
    return(opt_t)
```

```python
def equil_output(th,t_1, T_i, f_i, n, grid):
    
    t = t_af(th,t_1, T_i, f_i, n, grid )
    e= 1 - k * s(th,t_1, T_i, f_i, n, grid )
    
    return(output(t,th,e))
```

# Plots with varying n

```python
import numpy as np

def generate_random_grids(grid, n_grids, size=None, seed=None):
    """
    Split an initial grid into n_grids disjoint random subgrids, using all values of grid.

    Parameters
    ----------
    grid : array-like
        Initial grid.
    n_grids : int
        Number of subgrids to generate.
    size : int or list of ints or None
        - If int: each subgrid gets this size, and any leftover values are added
          to the last subgrid.
        - If list: explicit minimum sizes for each subgrid, and any leftover values
          are added to the last subgrid.
        - If None: the grid is split randomly into n_grids nonempty parts.
    seed : int or None
        Random seed.

    Returns
    -------
    list of numpy arrays
        Disjoint random subgrids whose union is exactly the input grid.
    """
    rng = np.random.default_rng(seed)
    grid = np.array(grid)

    if n_grids < 1:
        raise ValueError("n_grids must be at least 1.")

    if len(grid) < n_grids:
        raise ValueError("Grid must have at least as many points as n_grids.")

    shuffled = rng.permutation(grid)

    if size is None:
        # Random nonempty split using all elements
        cut_points = sorted(
            rng.choice(np.arange(1, len(grid)), size=n_grids - 1, replace=False)
        )
        return [arr.copy() for arr in np.split(shuffled, cut_points)]

    if isinstance(size, int):
        if size < 0:
            raise ValueError("size must be nonnegative.")
        total_needed = n_grids * size
        if total_needed > len(grid):
            raise ValueError("Requested total size exceeds grid length.")

        grids = []
        start = 0
        for i in range(n_grids):
            end = start + size
            if i == n_grids - 1:
                end = len(grid)  # add all leftovers to the last subgrid
            grids.append(shuffled[start:end].copy())
            start += size
        return grids

    if isinstance(size, (list, tuple, np.ndarray)):
        if len(size) != n_grids:
            raise ValueError("If size is a list, it must have length n_grids.")
        if any(s < 0 for s in size):
            raise ValueError("Sizes must be nonnegative.")
        if sum(size) > len(grid):
            raise ValueError("Requested total size exceeds grid length.")

        grids = []
        start = 0
        for i, s in enumerate(size):
            end = start + s
            if i == n_grids - 1:
                end = len(grid)  # add all leftovers to the last subgrid
            grids.append(shuffled[start:end].copy())
            start += s
        return grids

    raise TypeError("size must be None, an int, or a list of ints.")
```

```python
max_output = []
all_outputs = {}

for n in range(1, 50):
    all_grids = generate_random_grids(uniform_grid, n_grids=n, seed=123)
    d = alpha * (1 - p(n)) - k * gamma * p(n)
    print(d)

    if d > 0:
        all_outputs[str(n)] = {}
        max_outputs_n = []

        for i in range(n):
            grid = np.sort(all_grids[i])
            all_outputs[str(n)][str(i)] = {}

            weight = 1 / len(grid)

            def fi(x):
                return weight

            budget = T #* len(grid) / len(uniform_grid)
            maximal_value_t1 = max_t1(budget, fi, n, grid)

            if maximal_value_t1 <= 0:
                t1_values = [0.0]
            else:
                t1_values = [a / 100 for a in range(0, int(maximal_value_t1 * 100) + 1)]

            output_list = [
                sum(equil_output(th, t_1, budget, fi, n, grid) for th in grid)
                for t_1 in t1_values
            ]

            best_idx = int(np.argmax(output_list))
            best_output = output_list[best_idx]
            opt_t1 = t1_values[best_idx]

            max_outputs_n.append(best_output )

            all_outputs[str(n)][str(i)] = {
                'grid_no': i,
                'grid_values': grid.tolist(),
                'max_t1': maximal_value_t1,
                'all_output_values': list(zip(t1_values, output_list)),
                'max_output': best_output,
                'opt_t1': opt_t1,
                'opt_transfers': [
                    t_af(th, opt_t1, budget, fi, n, grid) for th in grid
                ],
            }

        max_output.append(sum(max_outputs_n))
```

```python
plt.plot([n for n in range(1, len(max_output)+1)], [sum([[b for a,b in all_outputs[str(n)][str(i)]['all_output_values']][0] for i in all_outputs[str(n)].keys()])/len(uniform_grid)
                                                    for n in range(1, len(max_output)+1)], color = 'black', label = 'Fixed welfare')
plt.plot([n for n in range(1, len(max_output)+1)], [el/len(uniform_grid) for el in max_output], color = 'firebrick', label = 'Delegation - uniform draws')
plt.legend()
plt.savefig('increase_output.png')
```

```python
plt.plot([n for n in range(1, len(max_output)+1)], [np.mean([all_outputs[str(n)][str(i)]['opt_t1'] for i in all_outputs[str(n)].keys()])
                                                    for n in range(1, len(max_output)+1)], color = 'black', label = 'Optimal slope of affine transfers')
plt.legend()
plt.savefig('optimal_slope.png')
```

```python
def split_grid(grid, n):
    """
    Split `grid` into `n` contiguous sub-grids with sizes as even as possible.

    Example:
    split_grid([1,2,3,4,5,6,7], 3)
    -> [[1,2,3], [4,5], [6,7]]
    """
    if n <= 0:
        raise ValueError("n must be a strictly positive integer")

    m = len(grid)
    q, r = divmod(m, n)  # q = base size, r = number of chunks with one extra element

    list_of_grids = []
    start = 0

    for i in range(n):
        chunk_size = q + 1 if i < r else q
        list_of_grids.append(grid[start:start + chunk_size])
        start += chunk_size

    return list_of_grids
```

```python
max_output = []
all_outputs = {}

for n in range(1, 50):
    all_grids = split_grid(uniform_grid, n=n)
    d = alpha * (1 - p(n)) - k * gamma * p(n)
    print(d)

    if d > 0:
        all_outputs[str(n)] = {}
        max_outputs_n = []

        for i in range(n):
            grid = np.sort(all_grids[i])
            all_outputs[str(n)][str(i)] = {}

            weight = 1 / len(grid)

            def fi(x):
                return weight

            budget = T #* len(grid) / len(uniform_grid)
            maximal_value_t1 = max_t1(budget, fi, n, grid)

            if maximal_value_t1 <= 0:
                t1_values = [0.0]
            else:
                t1_values = [a / 100 for a in range(0, int(maximal_value_t1 * 100) + 1)]

            output_list = [
                sum(equil_output(th, t_1, budget, fi, n, grid) for th in grid)
                for t_1 in t1_values
            ]

            best_idx = int(np.argmax(output_list))
            best_output = output_list[best_idx]
            opt_t1 = t1_values[best_idx]

            max_outputs_n.append(best_output )

            all_outputs[str(n)][str(i)] = {
                'grid_no': i,
                'grid_values': grid.tolist(),
                'max_t1': maximal_value_t1,
                'all_output_values': list(zip(t1_values, output_list)),
                'max_output': best_output,
                'opt_t1': opt_t1,
                'opt_transfers': [
                    t_af(th, opt_t1, budget, fi, n, grid) for th in grid
                ],
            }

        max_output.append(sum(max_outputs_n))
```

```python
plt.plot([n for n in range(1, len(max_output)+1)], [sum([[b for a,b in all_outputs[str(n)][str(i)]['all_output_values']][0] for i in all_outputs[str(n)].keys()])/len(uniform_grid)
                                                    for n in range(1, len(max_output)+1)], color = 'black', label = 'Fixed welfare')
plt.plot([n for n in range(1, len(max_output)+1)], [el/len(uniform_grid) for el in max_output], color = 'firebrick', label = 'Delegation - splitting')
plt.legend()
```
